# データ可視化 - 円グラフ

```python
# read modle
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
```

全体的な割合が各々どれくらいあるのか見るときに使います。

円グラフを描くには`pie`を使って、それぞれのサイズやラベル等を設定します。`axis`でグラフを調整します。`autopct`でそれぞれの割合を指定します。また`explode`を指定すると、特定のカテゴリーだけ、円グラフの全体から離す調整ができます（ここでは、`Hogs`を0.1に設定しています）。

```python
plt.figure(figsize=(5,3),facecolor="white")

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0) 

# setting size
plt.figure(figsize=(15, 6))

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
```
